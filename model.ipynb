## 3. Heat Transfer Maps (Spatial Grid Visualization)

```python
temp_cols = [f'T_y{y}_x{x}' for y in range(5) for x in range(3)]
vmin = df[temp_cols].values.min() - 0.3
vmax = df[temp_cols].values.max() + 0.3

plot_rows = df.sample(6, random_state=1).reset_index(drop=True)

fig, axes = plt.subplots(2, 3, figsize=(14, 9))
fig.suptitle('LT3 Heat Transfer Maps (5x3 Vertex Grid)', fontsize=14, fontweight='bold')

for i, (_, row) in enumerate(plot_rows.iterrows()):
    ax = axes[i // 3][i % 3]
    grid_vals = np.array([row[f'T_y{y}_x{x}'] for y in range(5) for x in range(3)]).reshape(5, 3)
    im = ax.imshow(grid_vals, cmap='hot', aspect='auto', vmin=vmin, vmax=vmax)
    for y in range(5):
        for x in range(3):
            ax.text(x, y, f'{grid_vals[y,x]:.1f}', ha='center', va='center', fontsize=9,
                    color='white' if grid_vals[y,x] > (vmin+vmax)/2 else 'black', fontweight='bold')
    ax.set_title(f"{row['date'].strftime('%d %b')} {row['period']}\n{int(row['people'])} people | {int(row['windows_open'])} windows open", fontsize=9)
    ax.set_xlabel('x ->'); ax.set_ylabel('y')
    ax.set_xticks([0,1,2]); ax.set_yticks([0,1,2,3,4])
    plt.colorbar(im, ax=ax, shrink=0.8, label='°C')

plt.tight_layout()
plt.show()
```

## 5. ML Model — Predict Average Room Temperature

```python
features = ['windows_open', 'people', 'period_num', 'day_of_week', 'is_weekend']
target = 'avg_temp'

X = df[features]
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

models = {
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
    'Ridge Regression': Ridge(alpha=1.0),
}

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    cv = cross_val_score(model, X, y, cv=5, scoring='r2')
    results[name] = {
        'MAE': round(mean_absolute_error(y_test, y_pred), 3),
        'RMSE': round(np.sqrt(mean_squared_error(y_test, y_pred)), 3),
        'R2': round(r2_score(y_test, y_pred), 3),
        'CV_R2_mean': round(cv.mean(), 3),
        'model': model
    }
    print(f'{name}: MAE={results[name]["MAE"]}C | RMSE={results[name]["RMSE"]}C | R2={results[name]["R2"]} | CV R2={results[name]["CV_R2_mean"]}')

best_name = min(results, key=lambda k: results[k]['MAE'])
print(f'\nBest model by MAE: {best_name}')
```

## 6. Feature Importance & Actual vs Predicted

```python
rf = results['Random Forest']['model']
importances = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=True)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

importances.plot(kind='barh', ax=axes[0], color='#e76f51')
axes[0].set_title('Feature Importance (Random Forest)')
axes[0].set_xlabel('Importance score')

y_pred_all = rf.predict(X)
sc = axes[1].scatter(y, y_pred_all, alpha=0.75, c=df['period_num'], cmap='RdYlBu_r', s=70)
axes[1].plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=1.5, label='Perfect fit')
axes[1].set_xlabel('Actual Temp (C)')
axes[1].set_ylabel('Predicted Temp (C)')
axes[1].set_title('Actual vs Predicted (all data)')
axes[1].legend()
plt.colorbar(sc, ax=axes[1], label='Period (0=morning, 2=evening)')

plt.tight_layout()
plt.show()
```
